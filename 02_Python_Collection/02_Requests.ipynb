## quiz : 네이버 증권 주요뉴스의 제목과 url 을 모두 추출하기

- 제목 변수, url 변수 만들기

```python
# 제목과 url 데이터 목록 만들기

news_title = []
news_url = []

for i in srclist:
    title = i.text
    url = urljoin(PATH, i['href'])
    print(title, url)
    news_title.append(title)
    news_url.append(url)
```

## 데이터프레임 변환

```python
import pandas as pd
```

```python
df = pd.DataFrame({'제목':news_title, '주소':news_url}    )
df
```

```python
print(df)
```

## excel 파일 저장

```python
df.to_excel('output/naver.xlsx', index = False)
```

## 멜론차트, 노래제목과 가수 수집하기

```python
resp = requests.get(PATH, headers = info)
resp
```

```python
html_src = resp.text
html_src
```

```python
soup = BeautifulSoup(html_src, 'lxml')
soup
```

```python
# 곡이름 찾기
len(soup.select('.ellipsis.rank01 a'))
```

```python
song = [item.text for item in soup.select('.ellipsis.rank01 a')]
len(song)
```

```python
len(soup.select('div.ellipsis.rank02>a'))
```

```python
artist = []
for item in soup.select('div.ellipsis.rank02>a') :
    artist.append(item.text)
print(len(artist))
print(artist)
```

```python
len(soup.select('.checkEllipsis'))
```

```python
name = [item.text for item in soup.select('.checkEllipsis')]
len(name)
```

```python
artist = []
for item in soup.select('.checkEllipsis') :
    artist.append(item.text)
print(len(artist))
print(artist)
```

```python
# 인덱스 만들기
rank_list = list(range(1, 101))
rank_list
```

```python
# 판다스로 데이터프레임 만들기 column : data
songDF = pd.DataFrame({
        "순위":rank_list,
        "노래제목":song,
        "가수명":artist})
songDF
```

```python
songDF.to_excel('melon100.xlsx', index=False)
```
